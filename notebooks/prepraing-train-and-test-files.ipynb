```python
import os
import random


# buffalo
full_path_to_images = r'/content/drive/MyDrive/Projects/african-wildlife-detection/images/buffalo'
os.chdir(full_path_to_images)
p1 = []
for current_dir, dirs, files in os.walk('.'):
    for f in files:
        if f.endswith('.jpg'):
            path_to_save_into_txt_files = full_path_to_images+ "//" +f
            p1.append(path_to_save_into_txt_files+'\n')

p1_test = p1[:int(len(p1)*0.20)]
# print(p1_test)
p1_test = random.sample(p1_test,len(p1_test))
# print(p1_test)
p1 = p1[int(len(p1)*0.20):]
# print(p1)
p1 = random.sample(p1,len(p1))
# print(p1)

# elephant
full_path_to_images = r'/content/drive/MyDrive/Projects/african-wildlife-detection/images/elephant'
os.chdir(full_path_to_images)
p2 = []
for current_dir, dirs, files in os.walk('.'):
    for f in files:
        if f.endswith('.jpg'):
            path_to_save_into_txt_files = full_path_to_images+ "//" +f
            p2.append(path_to_save_into_txt_files+'\n')

p2_test = p2[:int(len(p2)*0.20)]
# print(p2_test)
p2_test = random.sample(p2_test,len(p2_test))
# print(p2_test)
p2 = p2[int(len(p2)*0.20):]
# print(p2)
p2 = random.sample(p2,len(p2))
# print(p2)

# rhino
full_path_to_images = r'/content/drive/MyDrive/Projects/african-wildlife-detection/images/rhino'
os.chdir(full_path_to_images)
p3 = []
for current_dir, dirs, files in os.walk('.'):
    for f in files:
        if f.endswith('.jpg'):
            path_to_save_into_txt_files = full_path_to_images+ "//" +f
            p3.append(path_to_save_into_txt_files+'\n')

p3_test = p3[:int(len(p3)*0.20)]
# print(p3_test)
p3_test = random.sample(p3_test,len(p3_test))
# print(p3_test)
p3 = p3[int(len(p3)*0.20):]
# print(p3)
p3 = random.sample(p3,len(p3))
# print(p3)

# zebra
full_path_to_images = r'/content/drive/MyDrive/Projects/african-wildlife-detection/images/zebra'
os.chdir(full_path_to_images)
p4 = []
for current_dir, dirs, files in os.walk('.'):
    for f in files:
        if f.endswith('.jpg'):
            path_to_save_into_txt_files = full_path_to_images+ "//" +f
            p4.append(path_to_save_into_txt_files+'\n')

p4_test = p4[:int(len(p4)*0.20)]
# print(p4_test)
p4_test = random.sample(p4_test,len(p4_test))
# print(p4_test)
p4 = p4[int(len(p4)*0.20):]
# print(p4)
p4 = random.sample(p4,len(p4))
# print(p4)

p = p1 + p2 + p3 + p4
p = random.sample(p, len(p))

p_test = p1_test + p2_test + p3_test + p4_test
p_test = random.sample(p_test, len(p_test))

with open(r'/content/drive/MyDrive/Projects/african-wildlife-detection/files/train.txt','w') as train_txt:
    for e in p:
        train_txt.write(e)
        
with open(r'/content/drive/MyDrive/Projects/african-wildlife-detection/files/test.txt','w') as test_txt:
    for e in p_test:
        test_txt.write(e)
        
```
